```python
import pandas as pd
import numpy as np
import csv
district_data=r'C:\Users\marks\OneDrive\Desktop\Python folder\schools_complete.csv'
students_data=r'C:\Users\marks\OneDrive\Desktop\Python folder\students_complete.csv'
district_df=pd.read_csv(district_data)
students_df=pd.read_csv(students_data)
```

```python
total_schools=len(district_df["name"].unique())
total_students=district_df["size"].sum()

total_budget=district_df["budget"].sum()

avg_math=students_df["math_score"].sum()/students_df["math_score"].count()
avg_math=avg_math.round(2)
avg_reading=students_df["reading_score"].sum()/students_df["reading_score"].count()
avg_reading=avg_reading.round(2)


math_pass=students_df.loc[(students_df["math_score"]>=59.5)]
reading_pass=students_df.loc[(students_df["reading_score"]>=59.5)]

math_percentage_pass=math_pass["math_score"].count()/students_df["math_score"].count()
math_percentage_pass=math_percentage_pass.round(2)
reading_percentage_pass=reading_pass["reading_score"].count()/students_df["reading_score"].count()
reading_percentage_pass=reading_percentage_pass.round(2)
overall_percentage_pass=(math_pass["math_score"].count()+reading_pass["reading_score"].count())/(students_df["math_score"].count()+students_df["reading_score"].count())
overall_percentage_pass=overall_percentage_pass.round(2)

summary_table=pd.DataFrame({"Total Schools": total_schools,
               "Total Students": total_students,
               "Total Budget": total_budget,
               "Average Math Score": avg_math,
               "Average Reading Score": avg_reading,
               "Math Pass Rate": math_percentage_pass,
               "Reading Pass Rate": reading_percentage_pass,
               "Overall Pass Rate": overall_percentage_pass}, index=[0])
summary_table.head()
```

```python
schools_df1=district_df
schools_df1=schools_df1.rename(columns={"name":"school"})
                                                              
schools_df2=students_df

schools_df2=schools_df2[["school", "reading_score", "math_score"]]

mergedfile=pd.merge(schools_df1, schools_df2, on="school", how="outer")

mergedfile['budget_perstudent']=mergedfile['budget']/mergedfile['size']
mergedfile['avg_math']=mergedfile['math_score']/mergedfile['size']
mergedfile['avg_reading']=mergedfile['reading_score']/mergedfile['size']

mergedfile['math_pass']=np.where(mergedfile['math_score']>=59.5, 1, 0)
mergedfile['reading_pass']=np.where(mergedfile['reading_score']>=59.5, 1, 0)


mergedfile_grouped=mergedfile.groupby(['school'])

school_type=mergedfile_grouped['type'].unique()
school_budget=mergedfile_grouped['budget'].unique()
total_students=mergedfile_grouped['size'].unique()
perstudent=mergedfile_grouped['budget_perstudent'].unique()
average_math=mergedfile_grouped['avg_math'].sum()
average_reading=mergedfile_grouped['avg_reading'].sum()
math_pass_percent=(mergedfile_grouped['math_pass'].sum()/mergedfile_grouped['size'].count())*100
reading_pass_percent=(mergedfile_grouped['reading_pass'].sum()/mergedfile_grouped['size'].count())*100
overall_pass_percent=(math_pass_percent+reading_pass_percent)/2

school_summary_table=pd.DataFrame({'Average Math Score': average_math,
                                  'Average Reading Score': average_reading,
                                  'Math Pass Rate': math_pass_percent,
                                  'Reading Pass Rate': reading_pass_percent,
                                  'Overall Pass Rate': overall_pass_percent,
                                  'School Type':school_type,
                                  'School Budget': school_budget,
                                  'Per Student Budget':perstudent,
                                  'Total Students':total_students})

school_summary_table
```

```python
bestschools_df=school_summary_table.sort_values(['Overall Pass Rate'], ascending=False)
bestschools_df.reset_index(inplace=True)
bestschools_df.head(8)
```

```python
worstschools_df=school_summary_table.sort_values(['Overall Pass Rate'], ascending=True)
worstschools_df.reset_index(inplace=True)
worstschools_df.head(5)
```
